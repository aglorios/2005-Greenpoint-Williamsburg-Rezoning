Greenpoint 2005 Rezoning: Putting a Price on the Zoning Changes

During his time in office, Mayor Bloomberg encouraged private investment in New York's real estate market by overhauling the zoning plans for 40% of the city. One of the largest zoning plans that he implemented was along the waterfront in Williamsburg/Greenpoint in 2005. During this time, the rules regulating what kinds of development could occur in the area changed drastically. In a matter of years, the manufacturing districts that claimed the regaion for decades were replaced by high-rise luxury buildings. In order to appease a community that feared displacement, the Mayor's administration, agreed to include affordable housing through of private incentives to developers. However, after the global recession of 2008, many of the residential towers that were to include affordable housing remained unbuilt. In 2014, some of these luxury towers are set to come to the waterfront in Greenpoint. 

However, this development will occur against the backdrop of a different economic and political landscape in New York. Today's Mayor, Bill de Blasio, is less worried about encouraging private investment and more worried about securing affordable housing. In his effort to ensure 200,000 units of affordable housing over the next three years, Mayor Bill de Blasio is looking at all areas of the city with "fresh eyes." Many neighborhoods who have failed to be developed since their rezonings during the past 10 years may be rezoned again. Greenpoint is one of those neighborhoods.

One of the ways to ensure affordable housing in a development project is to mandate "inclusionary housing," which means that for every 80 units that are market price, 20 units must be affordable. Unlike during the Bloomberg administration, the developers won't be given extra tax breaks or benefits in order to build these affordable units. In this analysis, I measure how much potential money developers earned from the zoning changes in 2005, in what I see as a first step to understanding whether mandatory inclusionary zoning is feasible for New York.

In order to do this analysis, I worked with Pluto data from 2005 and 2013. I ensure that I was working with only Greenpoint's data by importing the two datasets into QGIS, isolating Greenpoint, and then re-exporting the new data sets.

In Python, I cleaned the data, merged them on BBL, and then did a series of calculations in order to put a price value on the zoning changes. I then created a table showing the Zoning Changes, the Square Foot Changes, and the Prices Changes. I then mapped the price changes on two different plots, first showing the price increase by property in red, and then showing the price decrease by property in blue.

Next, I want to calculate the square foot changes for the rezoning. This is somewhat tricky. In general, in order to calculate the buildable square feet for a piece of property, you must multiply the total Floor Area Ratio (FAR) by the Lot Area. The FAR for a property respresents how high a developer can build for a kind of building. For example, if the lot area is 1000 and the Residential zoning has a FAR of 4, then the total square feet that can be built for residential units is 4,000 feet. 

I will need to calculate first, the total square feet that can be built for each property in 2005 and 2013, and second, the indivdual square feet that can be built for each zoning type on each piece of property in 2005 and 2013. While I will ultimately show the total square foot difference in my table, my price for that total will be caculated by each individual square foot difference, respective to each of its zoning components on the lot. It is very common for a single piece of property to have zoning requirements for manufacturing, residential, and commercial - each piece with a different height requirement and different price value that corresponds to its respective market demand.

So, to start, I will calculate the total buildable square feet for each lot in 2005 and 2013. 

The 2005 data set looks slightly different from the 2013 data set in that it does not break down zoning be different FARs. Instead, it simply has one Maximum Allowable Far ('MAXALLWFAR'). This fact makes it easier to calculate to the total buildable square feet 2005 but makes it harder to measure the zoning changes for 2013, which I will explain in more detail further into the analysis.

In order to calculate 2005 buildable square feet, I simply multiply the total FAR by the Lot Area:

```python
data2005["SQFAR05"] = data2005.MAXALLWFAR*data2005.LOTAREA
```

There is not Max FAR for 2013. Instead, there are individual residential, community facility, and commercial FARs. So, in order to calculate the total buildable square feet for 2013, I can add these FARs and multiply their totoal by the lot area. 

Note: there is a built FAR for 2013 but I don't think this FAR measures the potential to build by selling air rights, which was crucial in Greenpoint's 2005 rezoning and subsequent development.

```python
data2013FAR = data2013[['CommFAR','FacilFAR','ResidFAR']]

data2013FAR['sumFAR'] = data2013FAR.sum(axis=1)

data2013FAR
```

```python
data2013['MFAR13'] = data2013FAR['sumFAR']

data2013[['CommFAR', 'FacilFAR', 'ResidFAR', 'MFAR13']]
```

```python
data2013["SQFAR13"] = data2013.MFAR13*data2013.LotArea
data2013.columns
```

Next, I need to separate the zoning differences so that I can multiply the components by different price values. 

For now, I am going to multiply CommSQFT by 50, FacilSQFT by 50, and ResidSQFT by 200. I might need to adjust FacilSQFT. Again, I also need to look into why Manufacturing is not included in the calculation.

FARDiff = SQFAR13 - SQFAR05

= ((CommFAR + ResidFAR + FacilFAR)*LotArea)) - (MAXALLWFAR*LOTAREA)
= ((CommFAR*LotArea) + (ResidFAR*LotArea) + (FacilFAR*LotArea)) - (MAXALLWFAR*LOTAREA)
= ((CommSQFT) + (ResidSQFT) + (FacilSQFT)) - (MaxSQFT)
FARDiff + MaxSQFT = CommSQFT + ResidSQFT + FacilSQFT

1. Calculate CommSQFT13, ResidSQFT13, FacilSQFT13, FARDIFF+SQFAR05
2. Add them to databoth[]
3. Take the CommDiff, ResidDiff, and FacilDiff
4. Multiply those difference by their respective price constants.

```python
#1. Calculate CommSQFT13, ResidSQFT13, FacilSQFT13, FARDIFF+SQFAR05
#2. Add them to databoth[]

databoth['CommSQFT13'] = databoth['CommFAR']*databoth['LotArea']
databoth['FacilSQFT13'] = databoth['FacilFAR']*databoth['LotArea']
databoth['ResidSQFT13'] = databoth['ResidFAR']*databoth['LotArea']
databoth['CommSQFT05'] = databoth['COMMFAR05']*databoth['LOTAREA']
databoth['FacilSQFT05'] = databoth['FACILFAR05']*databoth['LOTAREA']
databoth['ResidSQFT05'] = databoth['RESIDFAR05']*databoth['LOTAREA']
```

```python
#3. Take the Commercial sqft difference, Residential sqft difference, and community facility sqft difference.
# Add them to databoth[]

databoth['CommDiff'] = databoth['CommSQFT13'] - databoth['CommSQFT05']
databoth['FacilDiff'] = databoth['FacilSQFT13']- databoth['FacilSQFT05']
databoth['ResidDiff'] = databoth['ResidSQFT13'] - databoth['ResidSQFT05']
```

```python
#4. Multiply those difference by their respective price constants.
# Store in databoth[]

databoth['CommPriceDiff'] = databoth['CommDiff']*50
databoth['FacilPriceDiff'] = databoth['FacilDiff']*50
databoth['ResidPriceDiff'] = databoth['ResidDiff']*200
```

Next, I will calculate the total Price Changes by adding the price changes for commercial, community facility, and residential. Then I will show the table:

```python
TotalPrice = databoth[['CommPriceDiff', 'FacilPriceDiff', 'ResidPriceDiff']]
TotalPrice['sum'] = TotalPrice.sum(axis=1)

databoth['Price Difference'] = TotalPrice['sum']

databoth['Square Foot Difference'] = databoth['FARDiff']

databoth[['Zoning Changes', 'Square Foot Difference', 'Price Difference']]
```

This is good but I want to show the cumulative changes for each zoning change. In order to do that, I will use the groupby function and group by zoning change:

```python
ndataboth1 = databoth[['Zoning Changes', 'Square Foot Difference', 'Price Difference']]

sgrouped = ndataboth1.groupby(['Zoning Changes'],sort = True) #sorts the data

sumprice = sgrouped.sum() #sums the changes using zoning changes as the reference point

sumprice.sort('Price Difference', ascending = False) # sort in descending order
```
